```python
import glob
import cv2
import time
from tqdm import tqdm
from shutil import copyfile
from pathlib import Path
```

```python
og_images_pth = "/Users/nicholasho/Documents/GoVertical/unisal/examples/DHF1K/*/images/*.jpg"
og_images_pth = "/Users/nicholasho/Documents/Programming/_ai/hd2s/data/DHF1K/validation/frames/0601/*.jpg"
mask_pth = "/Users/nicholasho/Documents/Programming/_ai/hd2s/output/BaseModel/HD2S_testDHF1K_demo/0601/frames/*.png"
video_name = './output.mp4'
```

```python
path = glob.glob(og_images_pth)
masks = glob.glob(mask_pth)
images = sorted(path)
masks = sorted(masks)
```

```python
image_size = (128, 192)
```

```python
frame = cv2.imread(images[0])

# Display what the default shape is 
height, width, layers = frame.shape
og_shape = frame.shape
mask_shape = cv2.imread(masks[0]).shape

print(og_shape)
print(mask_shape)


fourcc = cv2.VideoWriter_fourcc(*'MP4V')

video = cv2.VideoWriter(video_name, fourcc, 25, (width,height))

for i, img in tqdm(enumerate(images)):
    mask = cv2.imread(masks[i])
    
    # You need the following if the input OG image isn't 
    # the same size as the output mask
    mask = cv2.resize(
                    mask,
                    dsize=(og_shape[1], og_shape[0]),
                    interpolation=cv2.INTER_CUBIC,
                )
    
    mask = mask.astype(float)/255
    img = cv2.imread(img)
    res = cv2.multiply(mask, img,dtype=cv2.CV_8U)
    video.write(res)

cv2.destroyAllWindows()
video.release()    
```
